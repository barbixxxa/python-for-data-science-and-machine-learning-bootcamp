```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

## Basic Matplotlib Commands

```python
# FUNCTIONAL METHOD
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

## Creating Multiplots on Same Canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Matplotlib Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.8,0.8])
```
