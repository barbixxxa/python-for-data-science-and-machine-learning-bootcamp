```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

## Basic Matplotlib Commands

```python
# FUNCTIONAL METHOD
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

## Creating Multiplots on Same Canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Matplotlib Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,0.8,0.8])
axes2 = fig.add_axes([0.9,0.0,1,0.4])

axes1.plot(x,y)
axes1.set_title('Left Plot')

axes2.set_title('Right Plot')
axes2.plot(y,x)
```

# Part 2 (L43)

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.legend(loc=0)
```

```python
fig.savefig('my_picture.jpg')
```

# Part 3 (L44)

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#ff8C00', linewidth=3, alpha=0.5, ls='--')
ax.plot(y,x,color='blue', marker='o', markersize=15, markerfacecolor='red', markeredgewidth=3, markeredgecolor='green')
```

## Plot Range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', lw=2, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
